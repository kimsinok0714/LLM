##Invoke
* 동기 방식의 기본적인 호출 방법
* 단일 프롬프트에 대한 모델의 완성된 전체 응답을 한 번에 반환

```python

messages = [
   SystemMessage("당신은 여행 전문가입니다."),
   HumanMessage("안녕하세요? 한국의 대표적인 관광지 3군데를 추천해주세요"),
   AIMessage("안녕하세요! 한국의 대표적인 관광지로는 다음 세 곳을 추천드립니다:\n\n1. **경복궁** (서울) - 한국의 전통 궁궐 문화를 느낄 수 있는 역사적인 장소입니다.\n2. **부산 해운대 해수욕장** - 아름다운 해변과 현대적인 도시가 어우러진 관광 명소입니다.\n3. **제주도** - 자연경관과 다양한 액티비티로 유명한 섬입니다."),
   HumanMessage("그 중에서 가족 단위 여행객에게 가장 적합한 곳은 어디인가요?")
]

# AIMessage
ai_message = model.invoke(messages)

print(ai_message)
# print(ai_message.content)

```

##스트리밍

* LLM 모델의 완성된 전체 응답을 한 번에 반환하는 것이 아니라 모델의 응답을 생성하는 즉시 실시간으로 토큰 단위로 전달하면 LangChain이 여러 토큰을 묶어서 청크 단위로 실시간 스트리밍으로 출력하는 예시


```python
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    streaming=True,  # 스트리밍 모드 활성화
    openai_api_key=os.environ["OPENAI_API_KEY"])


messages = [
  SystemMessage("You are a helpful assistant."),
  HumanMessage("Hello, how are you doing?")
]


for chunk in model.stream(messages):  # AIMessageChunk
  print(chunk.content, end = "", flush=True)
```

##Batch
여러 입력을 한꺼번에 처리해야 하는 상황에서 배치 처리가 필요합니다.

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage


model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=os.environ["OPENAI_API_KEY"])


inputs = [
    [HumanMessage(content="고양이에 대한 짧은 설명을 써줘.")],
    [HumanMessage(content="태양은 왜 뜨거운가?")],
    [HumanMessage(content="AI의 장점을 한 문장으로 정리해줘.")]
]

results = model.batch(inputs)

for r in results:
    print(r.content)
```

##비동기 호출


```python
# 여러 요청을 동시에 보내는 예시 (병렬 처리)
import asyncio
from langchain_openai import ChatOpenAI


async def query(model, question):
    response = await model.ainvoke(question)
    return response.content


async def main():

    model = ChatOpenAI(
              model="gpt-4o-mini",
              temperature=0,
              openai_api_key=os.environ["OPENAI_API_KEY"])

    tasks = [
        query(model, "고양이에 대한 짧은 설명을 써줘."),
        query(model, "태양은 왜 뜨거운가?"),
        query(model, "AI의 장점을 한 문장으로 정리해줘."),
    ]

    # asyncio.gather()를 사용하면 비동기 요청이라도 “입력한 순서대로” 결과가 리스트에 담기는 것이 보장됩니다.
    results = await asyncio.gather(*tasks)

    for r in results:
        print(r)


await main()

```

#Prompt template

* PromptTemplate = LLM에게 전달할 텍스트를 동적 변수로 구성할 수 있는 템플릿
* 즉, 틀(template)을 만들어두고, 입력값(input variable)을 채워서 완전한 프롬프트를 생성할 수 있음




[예시] LangChain의 PromptTemplate을 사용해서 프롬프트 템플릿 생성

```python
from langchain_core.prompts import PromptTemplate

# 1. 템플릿 문자열을 정의
# 2. PromptTemplate 객체 생셩
prompt = PromptTemplate.from_template("""
당신은 전문 요리사입니다.
다음 요리의 레시피를 작성해 주세요.

요리명: {dish}

조건:
1. 재료와 분량을 상세히 작성
2. 조리 순서와 시간, 온도를 구체적으로 표시
3. 요리 팁과 변형 방법 포함

최종 레시피는 단계별로 명확하게 작성해주세요.
""")

# 3. format 메소드를 사용하여 최종적으로 프롬프트 생성
print(prompt.format(dish="감자탕"))
```

[예시] LangChain에서 유사도 기반 Few-shot Prompt를 자동으로 구성
* Few-Shot Prompt + 유사도 기반 예시 선택 + 프롬프트 생성 흐름을 구현한 예제
* 사용자가 입력한 요리 이름(예: “감자탕”)과 비슷한 예시 레시피를 벡터 검색으로 찾아서, 그 예시를 프롬프트에 자동 삽입하는 구조입니다.


```python
from langchain_core.prompts import PromptTemplate, FewShotPromptTemplate
from langchain_openai import OpenAIEmbeddings
from langchain.vectorstores import FAISS


# ------------------------------
# 1. 예시용 프롬프트 템플릿
# ------------------------------

example_prompt = PromptTemplate.from_template(
    "요리명: {dish}\n레시피: {recipe}"
)

# ------------------------------
# 2. Few-shot 프롬프트 템플릿
# ------------------------------

few_shot_prompt = FewShotPromptTemplate(
    example_prompt=example_prompt,                                # 각 예시를 표현하는 PromptTemplate
    examples=[],                                                  # 여기에 나중에 동적으로 예시를 선택해서 삽입
    prefix="다음은 요리 이름과 해당 요리의 레시피 예시입니다.",   # 프롬프트의 시작 부분
    suffix="요리명: {dish}\n레시피:",                             # 사용자의 실제 입력 부분
    input_variables=["dish"]
)

# ------------------------------
# 3. 예시
# ------------------------------

examples = [
    {"dish": "김치찌개", "recipe": "김치, 돼지고기, 두부를 넣고 끓인다."},
    {"dish": "된장찌개", "recipe": "된장, 애호박, 두부를 넣고 끓인다."},
    {"dish": "감자조림", "recipe": "감자, 간장, 설탕을 넣고 조린다."},
    {"dish": "불고기", "recipe": "소고기를 간장, 설탕, 마늘, 참기름에 재워 볶는다."},
    {"dish": "비빔밥", "recipe": "밥 위에 각종 나물, 고기, 계란, 고추장을 올려 비벼 먹는다."},
    {"dish": "잡채", "recipe": "당면, 채소, 고기를 간장과 설탕으로 볶아 섞는다."},
    {"dish": "순두부찌개", "recipe": "순두부, 고기, 조개, 고추가루를 넣고 끓인다."},
    {"dish": "갈비찜", "recipe": "소갈비를 간장, 설탕, 마늘, 대파와 함께 푹 삶는다."},
    {"dish": "떡볶이", "recipe": "떡, 어묵을 고추장, 설탕, 간장 양념으로 볶는다."},
    {"dish": "오징어볶음", "recipe": "오징어, 양파, 당근을 고추장과 설탕으로 볶는다."}
]

# ------------------------------
# 4. FAISS 벡터 저장소 생성
# ------------------------------

embedding_model = OpenAIEmbeddings(openai_api_key=os.environ["OPENAI_API_KEY"])

vectorstore = FAISS.from_texts(
    [ex["dish"] for ex in examples],
    embedding_model,
    metadatas=examples
)


# ------------------------------------------------------------------------------------------
# 5. 유사도 기반 검색으로 벡터 저장소에서 사용자 질의와 유사한 문서 2개를 검색
# ------------------------------------------------------------------------------------------

query = "감자탕"
similar = vectorstore.similarity_search(query, k=2)


# ------------------------------------------------------------
# 6. FewShotPromptTemplate에 선택된 예시를 동적으로 삽입
# ------------------------------------------------------------

selected_examples = [s.metadata for s in similar]
few_shot_prompt.examples = selected_examples

# ------------------------------------------------------------
# 7. 프롬프트 생성
# ------------------------------------------------------------

final_prompt = few_shot_prompt.format(dish="감자탕")

print(final_prompt)
```

```python

# ------------------------------
# 8. Chat 모델 호출
# ------------------------------

model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=os.environ["OPENAI_API_KEY"])

ai_message = model.invoke(final_prompt)


print(ai_message.content)
```

#Output parser

* 대규모 언어 모델(LLM)의 비정형 텍스트 형태의 출력을 애플리케이션에서 활용할 수 있는 구조화된 데이터 형식으로 자동 변환해주는 컴포넌트입니다.

* JSON, 리스트, Pydantic 모델 등으로의 변환 자동화

###PydanticOutputParser
* Pydantic 모델을 활용하여 LLM 출력을 JSON 구조로 정의하고 파싱합니다.
* 데이터 유효성 검사와 타입 안전성 제공합니다.

* LangChain에서 LLM에게 JSON등의 구조화된 데이터를 출력하도록  하려면 PydanticOutputParser보다 with_structured_ouput 사용을 권장합니다.



```python
from pydantic import BaseModel, Field

# ----------------------------------------------------------------------------------
# Pydantic 모델 정의
# 요리 재료 목록(ingredients)와 요리 순서 목록(steps)를 필드로 갖는 Receipe 클래스
# ----------------------------------------------------------------------------------

class Receipe(BaseModel):
  ingredients: list[str] = Field(description="요리 재료")
  steps: list[str] = Field(description="요리 순서")
```

```python

from langchain_core.output_parsers import PydanticOutputParser

# PydanticOutputParser 생성
output_parser = PydanticOutputParser(pydantic_object=Receipe)

format_instructions = output_parser.get_format_instructions()

print(format_instructions)  # LLM에게 JSON 스키마에 맞춰 감자탕이나 김치찌개 등 실제 요리 레시피를 JSON 형식으로 출력해 달라고 요청(강제)할 수 있습니다.
```

```python
# format_instructions을 프롬프트에 포함시켜 LLM이 이 형식에 맞는 응답을 반환하도록 합니다.

from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
  [
    ("system" , "사용자가 입력한 요리의 레시피를 생각해주세요.\n\n{format_instructions}"),
    ("human", "{dish}"),
    ("ai", "좋습니다! {dish} 레시피를 단계별로 알려드리겠습니다.")
  ]
)

# format_instructions을 프롬프트에 포함시켜 LLM이 이 형식에 맞는 응답을 출력하도록 강제합니다.
prompt_with_format_instructions = prompt.partial(format_instructions=format_instructions)

prompt_value = prompt_with_format_instructions.invoke({"dish": "카레"})

print(prompt_value)

# print(prompt_value.messages)

# print(prompt_value.messages[0])  # SystemMessage
```

###StrOuputParser

* LangChain의 가장 기본적인 출력 파서로, LLM이나 체인의 응답을 단순 문자열(String)으로 반환합니다




```python
from langchain_core.messages import AIMessage
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate


prompt = PromptTemplate.from_template("""
당신은 전문 요리사입니다.
다음 요리의 레시피를 작성해 주세요.

요리명: {dish}

조건:
1. 재료와 분량을 상세히 작성
2. 조리 순서와 시간, 온도를 구체적으로 표시
3. 요리 팁과 변형 방법 포함

최종 레시피는 단계별로 명확하게 작성해주세요.
""")

model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=os.environ["OPENAI_API_KEY"])

output_parser = StrOutputParser()

# 체인 구성
chain = prompt | model | output_parser

# 실행
ai_message = chain.invoke({"dish": "김치찌개"})

print(ai_message)
print(type(ai_message))

```

###with_structured_output

* with_structured_output은 Chat 모델에서만 지원됩니다.

* LangChain에서 LLM의 비정형 텍스트 형태의 출력을 구조화된 데이터로 변환할 때 PydanticOutputParser보다 with_structured_ouput 사용을 권장합니다.



```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from pydantic import BaseModel, Field
import pprint


# ----------------------------------------------------------------------------------
# Pydantic 모델 정의
# 요리 재료 목록(ingredients)와 요리 순서 목록(steps)를 필드로 갖는 Receipe 클래스
# ----------------------------------------------------------------------------------
class Receipe(BaseModel):
  ingredients: list[str] = Field(description="요리 재료")
  steps: list[str] = Field(description="요리 순서")


# ----------------------------
# ChatPromptTemplate 정의
# ----------------------------

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "사용자가 입력한 요리의 레시피를 생각해주세요."),
        ("human", "{dish}"),
        ("ai", "좋습니다! {dish} 레시피를 단계별로 알려드리겠습니다.")
    ]
)

# ----------------------------
# ChatOpenAI 모델 정의
# ----------------------------

model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=os.environ["OPENAI_API_KEY"]
)


# LangChain은 LLM에게 정의된 스키마에 따라 출력하도록 지시합니다.
# LLM이 정의된 스키마에 따라 구조화된 출력을 네이티브 기능으로 생성합니다.
# LangChain은 결과를 거의 파싱하지 않습니다.
chain = prompt | model.with_structured_output(Receipe)

receipe = chain.invoke({"dish": "김치찌개"})

pprint.pprint(receipe)
# print(receipe.ingredients)
# print(receipe.steps)

```

도구/함수 호출
* 최신 LLM은 특정 기능을 호출하는 방식으로 구조화된 데이터를 생성할 수 있습니다


```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
import json

model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=os.environ["OPENAI_API_KEY"]
)

# HumanMessage 객체 사용
messages = [HumanMessage(content="김치찌개 레시피를 JSON으로 만들어줘")]

# 함수 호출 지원
functions = [{
    "name": "create_recipe",
    "description": "요리 레시피를 JSON 구조로 생성",
    "parameters": {
        "type": "object",
        "properties": {
            "ingredients": {"type": "array", "items": {"type": "string"}},
            "steps": {"type": "array", "items": {"type": "string"}}
        },
        "required": ["ingredients", "steps"]
    }
}]

# 모델 호출 (invoke 사용)
response = model.invoke(
    messages,
    functions=functions,
    function_call={"name": "create_recipe"}  # 모델에게 특정 함수 호출 지시 (create_recipe)
)

# print(response)

arguments_str = response.additional_kwargs['function_call']['arguments']


# 2. 문자열 → Python dict 변환
recipe_dict = json.loads(arguments_str)

# 3. ingredients 출력
print(recipe_dict["ingredients"])
```
